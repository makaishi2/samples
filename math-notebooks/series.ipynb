```python
import numpy as np   # ← これを必ず最初に書く

def taylor_series_sin(x, n_terms):
    """sin(x) のテイラー展開 n_terms 項までの近似"""
    result = 0
    for n in range(n_terms):
        term = x**(n) / math.factorial(n)
        result += term
    return result

# x を NumPy 配列で用意
x = np.linspace(-2, 2, 100)

# 近似の項数のリスト
n_terms_list = [1, 2, 3, 4, 5, 6]

# 各項数についての近似を計算
approximations = [taylor_series_sin(x, n) for n in n_terms_list]

# ちゃんと配列が入っているか確認
for n, y in zip(n_terms_list, approximations):
    print(n, y.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Define the function and Taylor series
def taylor_series_sin(x, n_terms):
    result = 0
    for n in range(n_terms):
        term = x**n / math.factorial(n)
        result += term
    return result

# Define the x range
x = np.linspace(-4, 4, 500)

# Calculate the sine function
y_sin = np.exp(x)

# Taylor series approximations
n_terms_list = [2, 3, 4]
approximations = [taylor_series_sin(x, n) for n in n_terms_list]
```

```python
# Plotting again with y-axis limited to [-2, 2]
plt.figure(figsize=(6, 6))

# Plot the actual sine function
plt.plot(x, y_sin, label="exp(x)", color="black", linewidth=2)

# Plot the Taylor series approximations
for n, approximation in zip(n_terms_list, approximations):
    plt.plot(x, approximation, label=f"{n-1}次近似", linewidth=1)

# Add labels, legend, and grid
#plt.title("sin(x)とテイラー級数")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-10,10)
plt.axhline(0, color="gray", linewidth=0.5, linestyle="--")
plt.axvline(0, color="gray", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python
# Define the Fourier series for a square wave
def fourier_series_square_wave(x, n_terms):
    result = np.zeros_like(x)
    for n in range(1, n_terms + 1, 2):  # Only odd terms
        result += (4 / (np.pi * n)) * np.sin(n * x)
    return result

# Define the x range
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Define the actual square wave function for comparison
def square_wave(x):
    return np.sign(np.sin(x))

y_square = square_wave(x)

# Fourier series approximations
n_terms_list = [1, 3, 5, 7, 9, 15]
approximations = [fourier_series_square_wave(x, n) for n in n_terms_list]

# Plotting
plt.figure(figsize=(12, 8))

# Plot the actual square wave
plt.plot(x, y_square, label="方形波", color="black", linewidth=2, linestyle="--")

# Plot the Fourier series approximations
for n, approximation in zip(n_terms_list, approximations):
    plt.plot(x, approximation, label=f"フーリエ級数 (n={n})", linewidth=0.8)

# Add labels, legend, and grid
plt.title("方形波とフーリエ級数")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-1.5, 1.5)
plt.xlim(-1, 4)
plt.axhline(0, color="gray", linewidth=0.5, linestyle="--")
plt.axvline(0, color="gray", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
