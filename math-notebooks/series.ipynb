```python
# 必要ライブラリのimport
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

# matplotlib日本語化対応
import japanize_matplotlib

# データフレーム表示用関数
from IPython.display import display

# グラフのデフォルトフォント指定
plt.rcParams["font.size"] = 14
```

```python
def taylor_series_exp(x, n_terms):
    """exp(x) のテイラー展開 n_terms 項までの近似"""
    result = 0
    for n in range(n_terms):
        term = x**(n) / math.factorial(n)
        result += term
    return result

# x を NumPy 配列で用意
x = np.linspace(-2, 2, 100)

# 近似の項数のリスト
n_terms_list = [1, 2, 3, 4, 5, 6]

# 各項数についての近似を計算
approximations = [taylor_series_exp(x, n) for n in n_terms_list]

y_exp = np.exp(x)

```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8), tight_layout=True)
for i in range(2*3):
    ax = axes[i//3, i%3]
    #ax.set_title(f'{i} 次近似')
    ax.plot(x, y_exp, label=r"$y=e^x$", color="black", linewidth=2)
    series = approximations[i]
    ax.plot(x, series, label=f"{i}次近似", linewidth=2, c='b')
    ax.legend()
    ax.grid()
create_pdf()
plt.show()
```
