```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# グラフ作成
fig, ax = plt.subplots(figsize=(5, 5))

ax.set_xlim(-2,8)
ax.set_ylim(-2,8)


# スパイン調整
ax.spines['left'].set_position('zero')   # y軸をx=0に設定
ax.spines['bottom'].set_position('zero')  # x軸をy=0に設定
ax.spines['right'].set_color('none')  # 右のスパインを非表示
ax.spines['top'].set_color('none')    # 上のスパインを非表示

# 軸の目盛りを調整
ax.xaxis.set_ticks_position('bottom')  # x軸の目盛りを下側に表示
ax.yaxis.set_ticks_position('left')    # y軸の目盛りを左側に表示

# y軸の目盛刻みを0.5に設定
x_major_locator = MultipleLocator(1)  # 目盛刻み幅0.5
ax.xaxis.set_major_locator(x_major_locator)

# y軸の目盛刻みを0.5に設定
y_major_locator = MultipleLocator(1)  # 目盛刻み幅0.5
ax.yaxis.set_major_locator(y_major_locator)

# 軸の目盛りのスタイル
ax.tick_params(axis='x', direction='inout', length=6)  # x軸の目盛り
ax.tick_params(axis='y', direction='inout', length=6)  # y軸の目盛り

# ベクトルを追加
ax.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1, color='black')
ax.quiver(0, 0, 0, 1, angles='xy', scale_units='xy', scale=1, color='black')

ax.quiver(0, 0, 2, 1, angles='xy', scale_units='xy', scale=1, color='blue')
ax.quiver(0, 0, 1, 3, angles='xy', scale_units='xy', scale=1, color='blue')

# 凡例の追加
#ax.legend(fontsize=14)

# ex
plt.text(0.2, -0.5, r'$\mathbf{e}_x$', fontsize=12)
# ey
plt.text(-0.8, 0.4, r'$\mathbf{e}_y$', fontsize=12)

#  T(ex)
plt.text(1.5, 0.2, r'$\mathbf{T}(\mathbf{e}_x)$', color='blue', fontsize=12)
# T(ey)
plt.text(0.5, 3.3, r'$\mathbf{T}(\mathbf{e}_y)$', color='blue', fontsize=12)

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')

plt.grid()
plt.show()
```
